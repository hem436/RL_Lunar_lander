# 🚀 Reinforcement Learning Agents for LunarLander

**Deep Deterministic Policy Gradient (DDPG)** and **Proximal Policy Optimization (PPO)** applied to the `LunarLanderContinuous-v3` environment from Gymnasium.

---
![ALT](./lunar_lander.gif)
### Algorithm Overview

| Feature | DDPG | PPO |
|---|---|---|
| **Policy type** | Deterministic | Stochastic (Gaussian) |
| **On/Off policy** | Off-policy | On-policy |
| **Action space** | Continuous | Continuous |
| **Key idea** | Actor-critic + replay buffer + target nets | Clipped surrogate objective |
| **Exploration** | Ornstein-Uhlenbeck noise | Policy entropy (σ head) |
| **Memory** | Experience replay buffer | Episode rollout buffer |

## 🧠 3. DDPG Architecture

```python
# ── Actor Network (from DDPG scratch: 400→200, no LayerNorm) ─────────────
class Actor(nn.Module):
    """Deterministic policy: state → action."""

    def __init__(self, state_dim, action_dim, hidden=(400, 200)):
        super().__init__()
        self.input_layer            = nn.Linear(state_dim, hidden[0])
        self.input_layer_activation = nn.ReLU()
        self.hidden_layer           = nn.Linear(hidden[0], hidden[1])
        self.output_layer           = nn.Linear(hidden[1], action_dim)
        self.output_activation      = nn.Tanh()   # outputs ∈ [-1, 1]
        self._init_weights()

    def _init_weights(self):
        for layer in [self.input_layer, self.hidden_layer, self.output_layer]:
            nn.init.xavier_uniform_(layer.weight)
            nn.init.zeros_(layer.bias)
        # Small uniform init for the final layer (stability)
        nn.init.uniform_(self.output_layer.weight, -3e-3, 3e-3)

    def forward(self, state):
        x = self.input_layer_activation(self.input_layer(state))
        x = self.input_layer_activation(self.hidden_layer(x))
        return self.output_activation(self.output_layer(x))


# ── Critic Network (from DDPG scratch: 400→200, action injected after layer 1)
class Critic(nn.Module):
    """Q-function: (state, action) → Q-value."""

    def __init__(self, state_dim, action_dim, hidden=(400, 200)):
        super().__init__()
        self.fc1  = nn.Linear(state_dim, hidden[0])
        self.fc2  = nn.Linear(hidden[0] + action_dim, hidden[1])
        self.fc3  = nn.Linear(hidden[1], 1)
        self._init_weights()

    def _init_weights(self):
        for layer in [self.fc1, self.fc2, self.fc3]:
            nn.init.xavier_uniform_(layer.weight)
            nn.init.zeros_(layer.bias)
        nn.init.uniform_(self.fc3.weight, -3e-3, 3e-3)

    def forward(self, state, action):
        x = F.relu(self.fc1(state))
        x = torch.cat([x, action], dim=1)
        x = F.relu(self.fc2(x))
        return self.fc3(x)


print('Actor  :', Actor(8, 2))
print('\nCritic :', Critic(8, 2))
```

```python
# ── Ornstein-Uhlenbeck Noise (parameters from DDPG scratch: mu=0.15, sigma=0.2)
class OUNoise:
    """Temporally correlated noise for exploration."""

    def __init__(self, size, mu=0.15, sigma=0.2):
        self.size  = size
        self.mu    = mu          # mean-reversion rate  (was theta in PIML)
        self.sigma = sigma       # noise std
        self.reset()

    def reset(self):
        self.state = np.zeros(self.size)

    def sample(self):
        # dx = -mu * x + N(0, sigma)  — matches DDPG-scratch convention
        dx           = -self.mu * self.state + np.random.normal(0, self.sigma, self.size)
        self.state  += dx
        return self.state.copy()


# ── Replay Buffer ─────────────────────────────────────────────────────
Transition = namedtuple('Transition',
                        ('state', 'action', 'reward', 'next_state', 'done'))

class ReplayBuffer:
    def __init__(self, capacity=50_000):      # L=50 000 from DDPG scratch
        self.buffer = deque(maxlen=capacity)

    def push(self, *args):
        self.buffer.append(Transition(*args))

    def sample(self, batch_size):
        batch = random.sample(self.buffer, batch_size)
        return Transition(*zip(*batch))

    def __len__(self):
        return len(self.buffer)


print('✅ OUNoise & ReplayBuffer defined.')
```

```python
# ── DDPG Agent (hyperparameters aligned with DDPG-from-Scratch notebook) ─
class DDPGAgent:
    def __init__(self,
                 state_dim, action_dim,
                 lr_actor    = 5e-5,    # was 1e-4  → from DDPG scratch
                 lr_critic   = 5e-4,    # was 1e-3  → from DDPG scratch
                 gamma       = 0.99,
                 tau         = 1e-3,    # was 0.005 → from DDPG scratch
                 buffer_size = 50_000,  # was 1e6   → from DDPG scratch (L=50 000)
                 batch_size  = 64,      # was 256   → from DDPG scratch (N=64)
                 actor_update_freq = 2):# d=2 from DDPG scratch

        self.gamma              = gamma
        self.tau                = tau
        self.batch_size         = batch_size
        self.actor_update_freq  = actor_update_freq
        self._update_step       = 0

        # ── Networks ─────────────────────────────────────────────────
        self.actor        = Actor(state_dim, action_dim).to(DEVICE)
        self.actor_target = copy.deepcopy(self.actor)

        self.critic        = Critic(state_dim, action_dim).to(DEVICE)
        self.critic_target = copy.deepcopy(self.critic)

        self.actor_opt  = optim.Adam(self.actor.parameters(),  lr=lr_actor)
        self.critic_opt = optim.Adam(self.critic.parameters(), lr=lr_critic)

        # ── Replay & Noise ────────────────────────────────────────────
        self.buffer = ReplayBuffer(buffer_size)
        self.noise  = OUNoise(action_dim)          # mu=0.15, sigma=0.2

        # ── Logging ───────────────────────────────────────────────────
        self.actor_losses  = []
        self.critic_losses = []

    # ──────────────────────────────────────────────────────────────────
    def select_action(self, state, add_noise=True, noise_scale=1.0):
        state_t = torch.FloatTensor(state).unsqueeze(0).to(DEVICE)
        self.actor.eval()
        with torch.no_grad():
            action = self.actor(state_t).cpu().numpy()[0]
        self.actor.train()
        if add_noise:
            action += noise_scale * self.noise.sample()
        return np.clip(action, -1, 1)

    # ──────────────────────────────────────────────────────────────────
    def update(self):
        if len(self.buffer) < self.batch_size:
            return

        self._update_step += 1

        batch       = self.buffer.sample(self.batch_size)
        states      = torch.FloatTensor(np.array(batch.state)).to(DEVICE)
        actions     = torch.FloatTensor(np.array(batch.action)).to(DEVICE)
        rewards     = torch.FloatTensor(np.array(batch.reward)).unsqueeze(1).to(DEVICE)
        next_states = torch.FloatTensor(np.array(batch.next_state)).to(DEVICE)
        dones       = torch.FloatTensor(np.array(batch.done)).unsqueeze(1).to(DEVICE)

        # ── Critic update ─────────────────────────────────────────────
        self.target_critic_eval()
        # self.target_actor_eval()
        with torch.no_grad():
            next_actions = self.actor_target(next_states)
            target_q     = self.critic_target(next_states, next_actions)
            target_q     = rewards + self.gamma * (1 - dones) * target_q

        self.actor.train(); self.critic.train()
        current_q   = self.critic(states, actions)
        critic_loss = F.mse_loss(current_q, target_q)

        self.critic_opt.zero_grad()
        critic_loss.backward()
        nn.utils.clip_grad_norm_(self.critic.parameters(), 1.0)
        self.critic_opt.step()

        # ── Actor update every d steps (d=2 from DDPG scratch) ────────
        if self._update_step % self.actor_update_freq == 0:
            actor_loss = -self.critic(states, self.actor(states)).mean()

            self.actor_opt.zero_grad()
            actor_loss.backward()
            nn.utils.clip_grad_norm_(self.actor.parameters(), 1.0)
            self.actor_opt.step()

            # ── Soft target updates ────────────────────────────────────
            self._soft_update(self.actor,  self.actor_target)
            self._soft_update(self.critic, self.critic_target)

            self.actor_losses.append(actor_loss.item())

        self.critic_losses.append(critic_loss.item())

    # ──────────────────────────────────────────────────────────────────
    def _soft_update(self, net, target_net):
        for p, tp in zip(net.parameters(), target_net.parameters()):
            tp.data.copy_(self.tau * p.data + (1 - self.tau) * tp.data)

    def target_critic_eval(self):
        self.critic_target.eval()
        self.actor_target.eval()

    # ──────────────────────────────────────────────────────────────────
    def save(self, path='ddpg_lunarlander.pt'):
        torch.save({
            'actor':  self.actor.state_dict(),
            'critic': self.critic.state_dict(),
        }, path)

    def load(self, path='ddpg_lunarlander.pt'):
        ckpt = torch.load(path, map_location=DEVICE)
        self.actor.load_state_dict(ckpt['actor'])
        self.critic.load_state_dict(ckpt['critic'])


print('✅ DDPGAgent defined.')
```

```python
# ── Capture frames from random policy ────────────────────────────────
def capture_episode(policy_fn, env_name='LunarLanderContinuous-v3',
                    max_steps=400, seed=0):
    """Run one episode and return (frames, total_reward)."""
    env_r = gym.make(env_name, render_mode='rgb_array')
    obs, _ = env_r.reset(seed=seed)
    frames, total_reward = [], 0.0
    for _ in range(max_steps):
        frame = env_r.render()
        frames.append(frame)
        action = policy_fn(obs)
        obs, reward, terminated, truncated, _ = env_r.step(action)
        total_reward += reward
        if terminated or truncated:
            frames.append(env_r.render())
            break
    env_r.close()
    return frames, total_reward


def display_video(frames, path='video.mp4', fps=30):
    """Save frames to an mp4 file and embed it in the notebook."""
    import base64
    from IPython.display import HTML, display
    imageio.mimsave(path, frames, fps=fps, format='mp4', macro_block_size=None)
    video = open(path, 'rb').read()
    b64 = base64.b64encode(video).decode('ascii')
    display(HTML(f'''
    <video width="640" height="480" controls>
        <source src="data:video/mp4;base64,{b64}" type="video/mp4">
    </video>
    '''))


# Random policy
random_env = gym.make('LunarLanderContinuous-v3')
random_frames, random_reward = capture_episode(
    lambda obs: random_env.action_space.sample(), seed=42)
random_env.close()

display_video(random_frames, './before_training.mp4', fps=25)
print(f'Random policy episode reward: {random_reward:.2f}')
```

## 🏋️ 6. Training the DDPG Agent

```python
# ── Hyperparameters (aligned with DDPG-from-Scratch notebook) ───────────
HP = dict(
    max_episodes       = 300,     # N_episodes=300  (was 1000)
    max_steps          = 1000,
    lr_actor           = 5e-5,    # from DDPG scratch (was 1e-4)
    lr_critic          = 5e-4,    # from DDPG scratch (was 1e-3)
    gamma              = 0.99,
    tau                = 1e-3,    # from DDPG scratch (was 0.005)
    batch_size         = 64,      # N=64 from DDPG scratch (was 256)
    buffer_size        = 50_000,  # L=50 000 from DDPG scratch (was 1e6)
    actor_update_freq  = 2,       # d=2 from DDPG scratch
    warmup_steps       = 5000,    # pure random exploration before learning
    noise_decay        = 0.9995,  # multiplicative noise scale decay
    noise_min          = 0.10,
    update_every       = 1,       # update every step (d handled inside agent)
    save_every         = 100,
    eval_every         = 50,
    eval_episodes      = 3,
    solved_reward      = 200,
    # OU noise params (from DDPG scratch)
    ou_mu              = 0.15,
    ou_sigma           = 0.2,
)

print('Hyperparameters:')
for k, v in HP.items():
    print(f'  {k:22s}: {v}')
```

```python
# ── Training Loop ─────────────────────────────────────────────────────
train_env = gym.make('LunarLanderContinuous-v3')
agent = DDPGAgent(
    state_dim          = obs_dim,
    action_dim         = act_dim,
    lr_actor           = HP['lr_actor'],
    lr_critic          = HP['lr_critic'],
    gamma              = HP['gamma'],
    tau                = HP['tau'],
    buffer_size        = HP['buffer_size'],
    batch_size         = HP['batch_size'],
    actor_update_freq  = HP['actor_update_freq'],
)

# Override OUNoise with scratch-aligned parameters
agent.noise = OUNoise(act_dim, mu=HP['ou_mu'], sigma=HP['ou_sigma'])

# ── Pre-fill replay buffer with random experiences (like DDPG scratch) ──
print('Pre-filling replay buffer with random experiences...')
pre_env = gym.make('LunarLanderContinuous-v3')
pre_state, _ = pre_env.reset(seed=SEED)
filled = 0
while filled < HP['warmup_steps']:
    action = pre_env.action_space.sample()
    next_s, rew, term, trunc, _ = pre_env.step(action)
    agent.buffer.push(pre_state, action, rew, next_s, float(term or trunc))
    pre_state = next_s
    filled   += 1
    if term or trunc:
        pre_state, _ = pre_env.reset()
pre_env.close()
print(f'Buffer pre-filled with {len(agent.buffer)} experiences.')

# ── Logging ───────────────────────────────────────────────────────────
episode_rewards   = []
eval_rewards      = []
eval_episodes_log = []
steps_per_episode = []
noise_scales      = []

noise_scale  = 1.0
total_steps  = 0
best_eval    = -np.inf

print('Starting DDPG training...')
print('─' * 65)

pbar = tqdm(range(1, HP['max_episodes'] + 1), desc='Training')

for episode in pbar:
    state, _ = train_env.reset(seed=episode)
    agent.noise.reset()
    ep_reward = 0.0
    ep_steps  = 0

    for step in range(HP['max_steps']):
        total_steps += 1

        action = agent.select_action(state, add_noise=True,
                                     noise_scale=noise_scale)

        next_state, reward, terminated, truncated, _ = train_env.step(action)
        done = terminated or truncated

        agent.buffer.push(state, action, reward, next_state, float(done))

        # Update every step (actor update cadence is d=2 inside agent.update)
        agent.update()

        state      = next_state
        ep_reward += reward
        ep_steps  += 1

        if done:
            break

    # Decay noise
    noise_scale = max(HP['noise_min'], noise_scale * HP['noise_decay'])

    episode_rewards.append(ep_reward)
    steps_per_episode.append(ep_steps)
    noise_scales.append(noise_scale)

    # ── Periodic evaluation ───────────────────────────────────────────
    if episode % HP['eval_every'] == 0:
        eval_env = gym.make('LunarLanderContinuous-v3')
        er = []
        for e_ep in range(HP['eval_episodes']):
            s, _ = eval_env.reset(seed=1000 + e_ep)
            done  = False
            r_sum = 0.0
            for _ in range(HP['max_steps']):
                a = agent.select_action(s, add_noise=False)
                s, r, term, trunc, _ = eval_env.step(a)
                r_sum += r
                if term or trunc:
                    break
            er.append(r_sum)
        eval_env.close()

        mean_eval = np.mean(er)
        eval_rewards.append(mean_eval)
        eval_episodes_log.append(episode)

        if mean_eval > best_eval:
            best_eval = mean_eval
            agent.save('./ddpg_best.pt')

    # ── Checkpoint ───────────────────────────────────────────────────
    if episode % HP['save_every'] == 0:
        agent.save(f'./ddpg_ep{episode}.pt')

    # ── Progress bar ─────────────────────────────────────────────────
    avg50 = np.mean(episode_rewards[-50:]) if len(episode_rewards) >= 50 \
            else np.mean(episode_rewards)
    pbar.set_postfix({
        'ep_reward': f'{ep_reward:.1f}',
        'avg50':     f'{avg50:.1f}',
        'best_eval': f'{best_eval:.1f}',
        'noise':     f'{noise_scale:.3f}',
        'buffer':    len(agent.buffer),
    })

    # Early stop if solved
    if avg50 >= HP['solved_reward'] and episode >= 50:
        print(f'\n🎉 Solved at episode {episode}! Avg50 = {avg50:.2f}')
        agent.save('./ddpg_solved.pt')
        break

train_env.close()
print('\n✅ Training complete.')
print(f'   Best eval reward : {best_eval:.2f}')
print(f'   Total steps      : {total_steps:,}')
```

## 🧠 12. PPO Architecture

```python
# ── PPO Actor Network (stochastic, dual-head: mean + sigma) ──────────
class PPOActor(nn.Module):
    """Stochastic policy: state → (mean, sigma) of Gaussian action dist."""

    def __init__(self, state_dim, action_dim, hidden=(400, 200)):
        super().__init__()
        # Shared trunk
        self.shared = nn.Sequential(
            nn.Linear(state_dim, hidden[0]),
            nn.ReLU(),
        )
        # Mean head  — tanh keeps outputs in [-1, 1]
        self.mean_hidden = nn.Linear(hidden[0], hidden[1])
        self.mean_out    = nn.Linear(hidden[1], action_dim)
        self.mean_act    = nn.Tanh()

        # Sigma head — sigmoid maps to (0, 1) for bounded std
        self.sigma_hidden = nn.Linear(hidden[0], hidden[1])
        self.sigma_out    = nn.Linear(hidden[1], action_dim)
        self.sigma_act    = nn.Sigmoid()

        self._init_weights()

    def _init_weights(self):
        for layer in [self.mean_hidden, self.mean_out,
                      self.sigma_hidden, self.sigma_out]:
            nn.init.xavier_uniform_(layer.weight)
            nn.init.zeros_(layer.bias)
        nn.init.uniform_(self.mean_out.weight,  -3e-3, 3e-3)
        nn.init.uniform_(self.sigma_out.weight, -3e-3, 3e-3)

    def forward(self, state):
        shared = self.shared(state)
        mean   = self.mean_act(self.mean_out(F.relu(self.mean_hidden(shared))))
        sigma  = self.sigma_act(self.sigma_out(F.relu(self.sigma_hidden(shared))))
        return mean, sigma


# ── PPO Critic Network (state → V(s)) ────────────────────────────────
class PPOCritic(nn.Module):
    """Value function: state → V(s)."""

    def __init__(self, state_dim, hidden=(400, 200)):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(state_dim, hidden[0]), nn.ReLU(),
            nn.Linear(hidden[0], hidden[1]), nn.ReLU(),
            nn.Linear(hidden[1], 1),
        )
        for layer in self.net:
            if isinstance(layer, nn.Linear):
                nn.init.xavier_uniform_(layer.weight)
                nn.init.zeros_(layer.bias)

    def forward(self, state):
        return self.net(state)


print("PPOActor  :", PPOActor(8, 2))
print("\nPPOCritic :", PPOCritic(8))
```

```python
# ── PPO Helper: Gaussian log-probability ─────────────────────────────
def gaussian_log_prob(actions, mean, sigma):
    """Compute log π(a|s) under independent Gaussian for each action dim."""
    dist = torch.distributions.Normal(mean, sigma + 1e-8)
    return dist.log_prob(actions).sum(dim=-1)          # sum over action dims


# ── PPO Agent ─────────────────────────────────────────────────────────
class PPOAgent:
    def __init__(self,
                 state_dim, action_dim,
                 lr_actor   = 1e-5,    # from PPO scratch notebook
                 lr_critic  = 1e-3,    # from PPO scratch notebook
                 gamma      = 0.99,
                 epsilon    = 0.2,     # PPO clip ratio
                 M          = 10):     # gradient epochs per rollout

        self.gamma   = gamma
        self.epsilon = epsilon
        self.M       = M

        self.actor  = PPOActor(state_dim, action_dim).to(DEVICE)
        self.critic = PPOCritic(state_dim).to(DEVICE)

        self.actor_opt  = optim.Adam(self.actor.parameters(),  lr=lr_actor)
        self.critic_opt = optim.Adam(self.critic.parameters(), lr=lr_critic)

        self.actor_losses  = []
        self.critic_losses = []

    # ──────────────────────────────────────────────────────────────────
    def select_action(self, state):
        """Sample action from Gaussian policy; return (action, log_prob)."""
        s = torch.FloatTensor(state).unsqueeze(0).to(DEVICE)
        with torch.no_grad():
            mean, sigma = self.actor(s)
        dist   = torch.distributions.Normal(mean, sigma + 1e-8)
        action = dist.sample()
        action = torch.clamp(action, -1, 1)
        return action.cpu().numpy()[0], mean.cpu().numpy()[0], sigma.cpu().numpy()[0]

    # ──────────────────────────────────────────────────────────────────
    def update(self, states_np, actions_np, returns_np):
        """Run M PPO epochs over the collected rollout."""
        states  = torch.FloatTensor(states_np).to(DEVICE)
        actions = torch.FloatTensor(actions_np).to(DEVICE)
        returns = torch.FloatTensor(returns_np).to(DEVICE)

        # Compute advantage with current critic (detached)
        with torch.no_grad():
            V_old   = self.critic(states).squeeze()
            Psi     = returns - V_old              # advantage estimate

        # old log-probs (fixed across epochs)
        with torch.no_grad():
            mean_old, sigma_old = self.actor(states)
            log_pi_old = gaussian_log_prob(actions, mean_old, sigma_old)

        for _ in range(self.M):
            # ── Critic update ─────────────────────────────────────────
            V = self.critic(states).squeeze()
            critic_loss = F.mse_loss(V, returns)

            self.critic_opt.zero_grad()
            critic_loss.backward()
            nn.utils.clip_grad_norm_(self.critic.parameters(), 1.0)
            self.critic_opt.step()

            # ── Actor update (PPO-clip) ───────────────────────────────
            mean, sigma = self.actor(states)
            log_pi      = gaussian_log_prob(actions, mean, sigma)
            ratio       = torch.exp(log_pi - log_pi_old)

            clip_ratio  = torch.clamp(ratio, 1 - self.epsilon, 1 + self.epsilon)
            actor_loss  = -torch.mean(torch.min(ratio * Psi, clip_ratio * Psi))

            self.actor_opt.zero_grad()
            actor_loss.backward()
            nn.utils.clip_grad_norm_(self.actor.parameters(), 1.0)
            self.actor_opt.step()

            self.actor_losses.append(actor_loss.item())
            self.critic_losses.append(critic_loss.item())

    # ──────────────────────────────────────────────────────────────────
    def save(self, path="ppo_lunarlander.pt"):
        torch.save({"actor": self.actor.state_dict(),
                    "critic": self.critic.state_dict()}, path)

    def load(self, path="ppo_lunarlander.pt"):
        ckpt = torch.load(path, map_location=DEVICE)
        self.actor.load_state_dict(ckpt["actor"])
        self.critic.load_state_dict(ckpt["critic"])


print("✅ PPOAgent defined.")
```

## 🏋️ 13. Training the PPO Agent

```python
# ── PPO Hyperparameters (from PPO scratch notebook) ──────────────────
PPO_HP = dict(
    max_episodes = 1600,   # N_episodes=1600  from PPO scratch
    max_steps    = 1000,
    lr_actor     = 1e-5,   # from PPO scratch
    lr_critic    = 1e-3,   # from PPO scratch
    gamma        = 0.99,
    epsilon      = 0.2,    # PPO clip ratio
    M            = 10,     # gradient epochs per rollout
    solved_reward= 200,
    eval_every   = 100,
    eval_episodes= 3,
    save_every   = 400,
    n_ep_running_average = 50,
)

print("PPO Hyperparameters:")
for k, v in PPO_HP.items():
    print(f"  {k:25s}: {v}")
```

```python
# ── PPO Training Loop ────────────────────────────────────────────────
ppo_train_env = gym.make("LunarLanderContinuous-v3")
ppo_agent = PPOAgent(
    state_dim  = obs_dim,
    action_dim = act_dim,
    lr_actor   = PPO_HP["lr_actor"],
    lr_critic  = PPO_HP["lr_critic"],
    gamma      = PPO_HP["gamma"],
    epsilon    = PPO_HP["epsilon"],
    M          = PPO_HP["M"],
)

ppo_episode_rewards   = []
ppo_eval_rewards      = []
ppo_eval_episodes_log = []
ppo_steps_per_episode = []

ppo_best_eval = -np.inf
print("Starting PPO training...")
print("─" * 65)

ppo_pbar = tqdm(range(1, PPO_HP["max_episodes"] + 1), desc="PPO Training")

for episode in ppo_pbar:
    state, _ = ppo_train_env.reset(seed=episode)
    done      = False
    ep_reward = 0.0

    # ── On-policy rollout buffers ─────────────────────────────────────
    ep_states   = []
    ep_actions  = []
    ep_rewards  = []

    t = 0
    while not done and t < PPO_HP["max_steps"]:
        action, _, _ = ppo_agent.select_action(state)
        next_state, reward, terminated, truncated, _ = ppo_train_env.step(action)
        done = terminated or truncated

        ep_states.append(state)
        ep_actions.append(action)
        ep_rewards.append(reward)

        state      = next_state
        ep_reward += reward
        t         += 1

    # ── Monte-Carlo returns G_t = Σ γ^k r_{t+k} ─────────────────────
    G, g = [], 0.0
    for r in reversed(ep_rewards):
        g = r + PPO_HP["gamma"] * g
        G.insert(0, g)

    # ── PPO update on full episode rollout ────────────────────────────
    ppo_agent.update(
        np.array(ep_states,  dtype=np.float32),
        np.array(ep_actions, dtype=np.float32),
        np.array(G,          dtype=np.float32),
    )

    ppo_episode_rewards.append(ep_reward)
    ppo_steps_per_episode.append(t)

    # ── Periodic evaluation ───────────────────────────────────────────
    if episode % PPO_HP["eval_every"] == 0:
        eval_env = gym.make("LunarLanderContinuous-v3")
        er = []
        for e_ep in range(PPO_HP["eval_episodes"]):
            s, _ = eval_env.reset(seed=1000 + e_ep)
            done_e, r_sum = False, 0.0
            for _ in range(PPO_HP["max_steps"]):
                a, _, _ = ppo_agent.select_action(s)
                s, r, term, trunc, _ = eval_env.step(a)
                r_sum += r
                if term or trunc:
                    break
            er.append(r_sum)
        eval_env.close()
        mean_eval = np.mean(er)
        ppo_eval_rewards.append(mean_eval)
        ppo_eval_episodes_log.append(episode)
        if mean_eval > ppo_best_eval:
            ppo_best_eval = mean_eval
            ppo_agent.save("./ppo_best.pt")

    if episode % PPO_HP["save_every"] == 0:
        ppo_agent.save(f"./ppo_ep{episode}.pt")

    avg50 = np.mean(ppo_episode_rewards[-50:]) if len(ppo_episode_rewards) >= 50             else np.mean(ppo_episode_rewards)
    ppo_pbar.set_postfix({
        "ep_reward": f"{ep_reward:.1f}",
        "avg50":     f"{avg50:.1f}",
        "best_eval": f"{ppo_best_eval:.1f}",
    })

    if avg50 >= PPO_HP["solved_reward"] and episode >= 50:
        print(f"\n🎉 PPO Solved at episode {episode}! Avg50 = {avg50:.2f}")
        ppo_agent.save("./ppo_solved.pt")
        break

ppo_train_env.close()
print("\n✅ PPO Training complete.")
print(f"   Best eval reward : {ppo_best_eval:.2f}")
```
